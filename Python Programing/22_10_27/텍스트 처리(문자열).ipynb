##### LAB 9-1 : 카이사르 암호를 만들어 보자

```python
import string

src_str = string.ascii_uppercase
dst_str = src_str[3:] + src_str[:3]

def ciper(a):          # 암호화 코드를 만드는 함소
    idx = src_str.index(a)
    return dst_str[idx]

src = input('문장을 입력하시오: ')
print('암호화된 문장 : ', end='')

for ch in src:
    if ch in src_str:
        print(ciper(ch), end='')
    else:
        print(ch, end='')

print()
```

##### 도전문제 9.3
- 트윗 데이터에서 대문자나 느낌표가 많이 나타나는 것은 글쓰는 사람의 감정이 흥분하거나 분노한 상태임을 나타내는 경우가 많다. 주어진 원천 트윗 데이터에서 대문자와 느낌표가 몇 번 사용되었는지 계산하는 코드를 작성해 보라. 느낌표가 몇 개인지 헤아리는 것은 count() 함수를 사용하면 쉽게 할 수 있다. 대문자는 어떻게 셀 수 있을까? 원 트윗을 list()를 이용하여 하나 하나의 문자로 분리하여 리스트를 만들 수 있다. 그리고 이 리스트의 각 항목 문자가 ch라고 할 때, ch.isupper()를 호출하면 대문자인 경우 True가 반환된다.

```python
t = "It's Not The Right Time To Conduct Exams. MY DEMAND IN BOLD AND CAPITAL. NO EXAMS IN COVID!!!"
l = t.lower()

t_lst = list(t)
print('느낌표 갯수 :', t_lst.count('!'))

count = 0
for ch in t_lst:
    if ch.isupper() == True:
        count += 1

print('대문자 갯수 :', count)
```

#####
메타문자 별표(*) : 앞 문자가 0개 이상 반복

```python
# AB* : A, AB, ABB, ABBB, ABBBB, ABBBB-----
print(re.search('AB*', 'A'))    # 'A' 조건에 맞음
print(re.search('AB*', 'AA'))    # 'A' 조건에 맞음
print(re.search('AB*', 'J-HOP'))   # 조건에 맞지 않음
print(re.search('AB*', 'X-MAN'))   # 'A' 조건에 맞음
print(re.search('AB*', 'CABBA'))      # 'ABB' 조건에 맞음
print(re.search('AB*', 'CABBBBBA'))      # 'ABBBBB' 조건에 맞음
```

#####
메타 문자 (?) : 직전에 있는 임의의 문자를 0회 또는 1회 반복한 패턴에 매치

```python
# AB? : A, AB
print(re.search('AB?', 'A'))    # 'A' 조건에 맞음
print(re.search('AB?', 'AA'))    # 'A' 조건에 맞음
print(re.search('AB?', 'J-HOP'))   # 조건에 맞지 않음
print(re.search('AB?', 'X-MAN'))   # 'A' 조건에 맞음
print(re.search('AB?', 'CABBA'))      # 'AB' 조건에 맞음
print(re.search('AB?', 'CABBBBBA'))      # 'AB' 조건에 맞음
```

#### LAB 9-7 : 패스워드 검사  

정규식을 이용한 패스워드 조건 검사 
1) 최소 8글자 이상  
2) 적어도 하나의 영문자, 대문자, 및 소문자를 포함해야 한다.  
3) 적어도 하나의 숫자를 포함해야 한다.  
4) 다음에 나타난 특수문자 [_, @, $, !] 중 하나를 반드시 포함해야 한다.  

```python
import re 

while True: 
    password = input("패스워드를 입력하세요 : "); 
    if len(password)<8 or not re.search("[a-z]", password) or \
        not re.search("[A-Z]", password) or \
        not re.search("[0-9]", password) or not re.search("[_@$!]", password): 
        print("유효하지 않은 패스워드!") 
    else: 
        print("유효한 패스워드") 
        break
```

```python
s = 'My lucky number 2 7 99'
re.sub('[0-9]+', '*', s)     # 숫자만 찾아서 *으로 바꿈
```

```python
re.sub('\d+', '*', s)     # 숫자만 찾아서 *으로 바꿈
```

교체함수(매치객체)  
re.sub('패턴', 교체함수, '문자열', 바꿀횟수)

```python
def multiple10(m):        # 매개변수로 매치 객체를 받음
    n = int(m.group())    # 매칭된 문자열을 가져와서 정수로 변환
    return str(n * 10)    # 숫자에 10을 곱한 뒤 문자열로 변환해서 반환

re.sub('[0-9]+', multiple10, '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
re.sub('[0-9]+', lambda m: str(int(m.group()) * 10), '1 2 Fizz 4 Buzz Fizz 7 8')
```

```python
s = 'My lucky number 2 7 99'

def hash_by_mult_and_modulo(m):        # 매개변수로 매치 객체를 받음
    n = int(m.group())    # 매칭된 문자열을 가져와서 정수로 변환
    print(n)
    return str(n * 23435 % 973)    # 숫자에 10을 곱한 뒤 문자열로 변환해서 반환

print(re.sub('[0-9]+', hash_by_mult_and_modulo, s))
```
