```python
import wikipedia
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# 위키백과 사전의 컨텐츠 제목을 명시해 준다
wiki = wikipedia.page('Artificial intelligence')
# 이 페이지의 텍스트 컨텐츠를 추출하도록 한다
text = wiki.content

# 워드 클라우드를 생성하기 위해 위의 코드를 삽입할 것
wordcloud = WordCloud(width = 2000, height = 1500).generate(text)
```

```python
plt.figure(figsize=(40,30))
# 화면에 이미지를 그려준다.
plt.imshow(wordcloud)
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
# 쓸데없는 언어들을 제외한 중지어를 출력한다.

s_words = STOPWORDS.union({'one', 'using', 'first','two','make','use', 'AI'})
wordcloud = WordCloud(width = 2000, height = 1500, stopwords = s_words).generate(text)
```

```python
plt.figure(figsize=(40,30))
# 화면에 이미지를 그려준다.
plt.imshow(wordcloud)
plt.show()
# 중요한 AI도 넣었는데 AI가 없어진 걸 알 수 있다.
```
